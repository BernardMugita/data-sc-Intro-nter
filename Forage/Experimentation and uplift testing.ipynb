```python
measure_calculations = df.groupby(['STORE_NBR', 'MONTH_ID']).agg(
    total_sales=('TOT_SALES', 'sum'),
    total_customers=('LYLTY_CARD_NBR', 'nunique'),
    txn_per_customer=('TXN_ID', 'count'),
    chips_per_customer=('PROD_QTY', 'sum'),
    avg_price_per_unit=('PACK_PRICE', 'mean')
).reset_index()

measure_calculations
```

```python
preTrialMeasures = measure_calculations[(measure_calculations['MONTH_ID'] <= '2019/02')]
preTrialMeasures
```

```python
control_stores = preTrialMeasures[(preTrialMeasures['STORE_NBR'] == 88) | (preTrialMeasures['STORE_NBR'] == 86) | (preTrialMeasures['STORE_NBR'] == 77)]
control_stores
```

```python
def calculate_correlation(input_table, metric_col, store_comparison):
    calc_corr_table = pd.DataFrame(columns=['Store1', 'Store2', 'corr_measure'])
    
    for store in input_table['STORE_NBR'].unique():
        if store != store_comparison:
            store1_measure = input_table[input_table['STORE_NBR'] == store][metric_col].values
            store2_measure = input_table[input_table['STORE_NBR'] == store_comparison][metric_col].values
            
            corr_measure = np.corrcoef(store1_measure, store2_measure)[0, 1]
            
            calc_corr_table = calc_corr_table.append({
                'Store1': store,
                'Store2': store_comparison,
                'corr_measure': corr_measure
            }, ignore_index=True)
    
    return calc_corr_table
```

```python
calculate_correlation(control_stores, 'total_sales', 88)
```

```python

# Filter the data for the period before the trial (YEARMONTH < 201903)
trial_store = preTrialMeasures['STORE_NBR']
control_store = control_stores['STORE_NBR']

# Add a Store_type column to label the trial store, control store, and other stores
preTrialMeasures['Store_type'] = preTrialMeasures['STORE_NBR'].apply(lambda x: 'Trial' if x == trial_store
                                                        else 'Control' if x == control_store
                                                        else 'Other stores')

# Calculate the mean total sales for each month and Store_type
past_sales = preTrialMeasures.groupby(['MONTH_ID', 'Store_type']).agg({'totSales': 'mean'}).reset_index()

# Convert YEARMONTH to a datetime format

# Plot the total sales by month with different colors for each Store_type
sns.lineplot(data=past_sales, x='MONTH_ID', y='total_Sales', hue='Store_type')
plt.xlabel('Month of operation')
plt.ylabel('Total sales')
plt.title('Total sales by month')
plt.show()
```
