```python
merger_one['time_race_result'].replace('\\N', pd.NaT, inplace=True)
merger_one['milliseconds_race_result'].replace('\\N', pd.NaT, inplace=True)
merger_one
```

```python
merger_one.dropna(subset=['time_race_result', 'milliseconds_race_result'], inplace=True)
merger_one
```

```python
merger_one['milliseconds_race_result'] = merger_one['milliseconds_race_result'].astype(str)
merger_one['position'] = merger_one['position'].astype(int)
merger_one['fastestLap'] = merger_one['fastestLap'].astype(int)
merger_one['rank'] = merger_one['rank'].astype(int)
```

```python
merger_one.dtypes
```

```python
milliseconds = merger_one['time_race_result']

def convert_to_datetime(race_time_str):
    try:
        hours, minutes, seconds = race_time_str.split(':')
        total_seconds = int(hours) * 3600 + int(minutes) * 60 + float(seconds)
        return pd.Timestamp.min + pd.Timedelta(seconds=total_seconds)
    except ValueError:
        return pd.NaT

# Convert the column to datetime using the custom function
merger_one['time_race_result'] = milliseconds.apply(convert_to_datetime)
```
