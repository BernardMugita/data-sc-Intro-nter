##### Quick Flashback :(Linear Regression) =>

```python
import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
def func(x):
    return slope * x + intercept
```

```python
model = list(map(func, x))
```

```python
plt.scatter(x, y)
plt.plot(x, model)
plt.show()
```

#### 3.0 Polynomial Regression

```python
# Example 2a:

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show()
```

```python
# Example 2b: Draw the line of Polynomial Regression-

import numpy
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

model = numpy.poly1d(numpy.polyfit(x, y, 3))  # * important

line = numpy.linspace(1, 22, 100)
plt.grid()
plt.scatter(x, y)
plt.plot(line,model(line))
plt.show()
```

```python
# Explanation of Example 2b:

# a.Import libraries:
import numpy
import matplotlib.pyplot as plt

# b.Create arrays that represent the values of the x and y axis:
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]


# c.We make a model using a method in Numpy:
model = numpy.poly1d(numpy.polyfit(x, y, 3))

# d.We specify how the line will be displayed:-
#   starting at position 1, and ending at position 22:
line = numpy.linspace(1, 22, 100)

# e. Draw the scatter plot :
plt.scatter(x, y)

# f. Draw the line of polynomial regression:
plt.plot(line, model(line))

# g. Display plot:
plt.show()

```

```python
# Example 3:
import numpy
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22] # => time in hours
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100] # => speed of car

model = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, model(x)))
```

```python
# Example 4:
# Predict the speed of a car passing at a tollbooth 20:00 :

from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22] # => time in(24 hour system)
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

model = numpy.poly1d(numpy.polyfit(x, y, 3))

speed = model(20)
print(speed)
```
