##### 4b. Creating An Age Column:

```python
### Creating the Age column:(For Age of Car)

age = []
launch_date = df['Latest_Launch']

for value in launch_date:
    year = int(value.split('/')[-1])
    age.append(2023 - year)
```

```python
df['Age']= age
```

```python
df.head()
```

##### 4c. Seperating  Manufacturers into class 1 and 2 :

```python
### Mean price per each Manufacturer 

mean_price_manufacturer = df[['Manufacturer', 'Price_in_thousands']].groupby('Manufacturer', as_index = False).mean()
mean_price_manufacturer
```

```python
### Seperating  Manufacturers into class 1 and 2

class_1 = []
class_2 = []

for index in range(len(mean_price_manufacturer)):
    if mean_price_manufacturer.iloc[index, 1] <= 30:
        class_1.append(mean_price_manufacturer.iloc[index, 0])
    else:
        class_2.append(mean_price_manufacturer.iloc[index, 0])
        
print('Manufacturers with less than 30 mean price: ', class_1)
print('-----------------------------------------------------------------------')
print('Manufacturers with more than 30 mean price: ', class_2)
```

```python
df.head()
```

##### 4d. Modify the  Manufacturers Column to  reflect class 1 and 2 :

```python
### Modifying the Manufacturer column in the df :

Manufacturer = df['Manufacturer']
new_Manufacturer = []

for value in Manufacturer:
    if value in class_1:
        new_Manufacturer.append(1)
    else:
        new_Manufacturer.append(2)
        
df['Manufacturer'] = new_Manufacturer
```
