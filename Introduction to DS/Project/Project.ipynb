```python
# A quick scan of the data revieles that there is an ID column in the three datasets
# the ID value is an identifier for all unique students in the three tables and is 
# therefore the fitting pivot for merging the three tables. 

math_physics = pd.merge(df1, df2, on='ID', how='inner')

math_physics.head(10)
```

```python
math_physics_data = pd.merge(math_physics, df3, on='ID')

math_physics_data.head(10)
```

```python
# Here I set my ID column as the Index column for perspective. 

math_physics_data = math_physics_data.set_index('ID')

math_physics_data.head(10)
```

```python
#  Here i replace the column label to label the columns appropriately

column_labels = {
    'Score_x': 'MathScore', 
    'Subject_x': 'Maths', 
    'Score_y': 'PhysicsScore',
    'Subject_y': 'Physics',
    'Score': 'DSScore',
    'Subject': 'Data Structures',
    }

math_physics_data = math_physics_data.rename(columns=column_labels)

math_physics_data.head(10)
```

```python
# Here i drop the misceleneous columns to make the data more refined. 

drop_columns = ['Age_x', 'Age_y', 'Sex_x', 'Sex_y', 'Ethinicity_x', 'Ethinicity_y', 'Maths', 'Physics', 'Data Structures']

math_physics_data.drop(columns=drop_columns, inplace=True)

math_physics_data.head(10)
```

```python
# Since binary or numerical data is suitable for data analysis, I will change my SEX 
# column to binary form. 

math_physics_data['Sex'] = math_physics_data['Sex'].replace({'M':1, 'F':2})

math_physics_data.head(10)
```

```python
# We can use this method to find the unique values in the entinicity column so that 
# we can know how diverser our student data is within that regard. 
math_physics_data['Ethinicity'].unique()
```

```python
# Now that we know which ethnicities our class comprises of we can replace them with
# numerical values like we did for 'Age'. 

math_physics_data['Ethinicity'] = math_physics_data['Ethinicity'].replace({
  'White American':1, 
  'European American':2, 
  'Hispanic':3,
  'African American':4  
})

math_physics_data.head(10)
```

```python
# Now our data is clean. We can show the summary statistic. 

math_physics_data.describe()
```

```python
# we can also use seaborn to visualize the score distribution for Sex

sns.countplot(data = math_physics_data, 
              x='MathScore', 
              hue='Sex'
              )
```
